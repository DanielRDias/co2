```python
import pandas as pd
import plotly.express as px

data=pd.read_csv('../input/co2-and-greenhouse-gas-emissions/owid-co2-data.csv')
data
```

```python
data_clean = data[(data['co2'] > 0) & (data['country'] != 'World')]
data_clean
```

```python
px.line(data_clean, x = 'year', y = 'co2', color='country')
```

```python
px.scatter(data_clean[data_clean['year']==2019], x="co2_per_capita", y="energy_per_capita",
           size="co2", color="country", hover_name="country", log_x=True, size_max=60)
```

```python
continent_data =  data[(data['country'].isin(['Europe', 'Africa', 'North America', 'South America', 'Oceania', 'Asia'])) & (data['co2'] > 0)]
continent_data
```

```python
px.line(continent_data, x = 'year', y = 'co2_per_capita', color='country')
```

```python
px.line(continent_data, x = 'year', y = 'co2', color='country')
```

```python
px.pie(continent_data, names='country', values='co2')
```
